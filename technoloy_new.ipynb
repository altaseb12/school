```python
from selenium import webdriver
from selenium.webdriver.chrome.service import Service
from webdriver_manager.chrome import ChromeDriverManager
from bs4 import BeautifulSoup
import time
import os
import pandas as pd
```

```python
# =========================
# 1. IMPORTS
# =========================
import requests
from bs4 import BeautifulSoup
import pandas as pd
import os
from urllib.request import urlretrieve
from tqdm import tqdm
import re
import time
from urllib.parse import urljoin
import zipfile

# =========================
# 2. SETTINGS
# =========================
BBC_BASE = "https://www.bbc.com/amharic/topics/c06gq8wx467t?page="
FANA_BASE = "https://www.fanamc.com/archives/category/tech/page/"
HEADERS = {"User-Agent": "Mozilla/5.0"}

TARGET_SIZE = 50   # 🔥 TEST SIZE

BASE_DIR = "test_multimodal_dataset"
IMG_DIR = os.path.join(BASE_DIR, "images")

os.makedirs(IMG_DIR, exist_ok=True)

data = []
visited = set()

# =========================
# 3. AMHARIC FILTER
# =========================
def is_amharic(text):
    return any('\u1200' <= c <= '\u137F' for c in text)

# =========================
# 4. CLEAN TEXT
# =========================
def clean_text(text):
    text = re.sub(r'\s+', ' ', text)
    return text.strip()

# =========================
# 5. EXTRACT TEXT
# =========================
def extract_text(soup, source):
    paragraphs = []

    if source == "BBC":
        content = soup.select("main p")
    else:
        content = soup.select("div.entry-content p")

    for p in content:
        t = p.get_text().strip()

        if len(t) > 30 and is_amharic(t):
            paragraphs.append(t)

    return clean_text(" ".join(paragraphs))

# =========================
# 6. EXTRACT IMAGES
# =========================
def extract_images(soup, base_url, source):
    images = []

    if source == "BBC":
        container = soup.select_one("main")
    else:
        container = soup.select_one("div.entry-content")

    if not container:
        return []

    img_tags = container.find_all("img")

    for img in img_tags:
        url = img.get("src") or img.get("data-src")

        if not url:
            continue

        full_url = urljoin(base_url, url)

        if any(x in full_url.lower() for x in ["logo", "icon", "avatar", "ads"]):
            continue

        if full_url not in images:
            images.append(full_url)

    return images[:5]

# =========================
# 7. SCRAPE ARTICLE
# =========================
def scrape_article(url, source):
    try:
        res = requests.get(url, headers=HEADERS, timeout=10)
        soup = BeautifulSoup(res.text, "html.parser")

        text = extract_text(soup, source)

        if len(text) < 100:
            return None

        title_tag = soup.find("h1")
        title = title_tag.get_text().strip() if title_tag else ""

        if not is_amharic(title):
            return None

        img_urls = extract_images(soup, url, source)

        if not img_urls:
            return None

        saved_paths = []

        for i, img_url in enumerate(img_urls):
            try:
                img_name = f"{len(data)}_{i}.jpg"
                path = os.path.join(IMG_DIR, img_name)

                urlretrieve(img_url, path)
                saved_paths.append(f"images/{img_name}")
            except:
                continue

        if not saved_paths:
            return None

        return {
            "title": title,
            "text": text,
            "source": source,
            "category": "Technology",   # ✅ NEW COLUMN
            "image_paths": ";".join(saved_paths)
        }

    except:
        return None

# =========================
# 8. SCRAPE BBC
# =========================
page = 1

while len(data) < TARGET_SIZE:
    print(f"\n📄 BBC Page {page} | Collected: {len(data)}")

    res = requests.get(BBC_BASE + str(page), headers=HEADERS)

    if res.status_code != 200:
        break

    soup = BeautifulSoup(res.text, "html.parser")

    links = soup.find_all("a")

    for a in tqdm(links):
        if len(data) >= TARGET_SIZE:
            break

        href = a.get("href")

        if href and "/amharic/" in href:
            if not href.startswith("http"):
                href = "https://www.bbc.com" + href

            if href in visited:
                continue
            visited.add(href)

            item = scrape_article(href, "BBC")

            if item:
                data.append(item)

    page += 1
    time.sleep(1)

# =========================
# 9. SCRAPE FANA
# =========================
page = 1

while len(data) < TARGET_SIZE:
    print(f"\n📄 FANA Page {page} | Collected: {len(data)}")

    res = requests.get(FANA_BASE + str(page), headers=HEADERS)

    if res.status_code != 200:
        break

    soup = BeautifulSoup(res.text, "html.parser")

    articles = soup.select("article")

    if not articles:
        break

    for art in tqdm(articles):
        if len(data) >= TARGET_SIZE:
            break

        try:
            a = art.find("a")
            link = a["href"]

            if link in visited:
                continue
            visited.add(link)

            item = scrape_article(link, "FANA")

            if item:
                data.append(item)

        except:
            continue

    page += 1
    time.sleep(1)

# =========================
# 10. SAVE CSV
# =========================
df = pd.DataFrame(data)

csv_path = os.path.join(BASE_DIR, "dataset.csv")

df.to_csv(csv_path, index=False, encoding="utf-8-sig")

print("\n✅ TOTAL COLLECTED:", len(data))

# =========================
# 11. ZIP DATASET
# =========================
zip_path = "test_multimodal_dataset.zip"

with zipfile.ZipFile(zip_path, 'w') as z:
    for root, dirs, files in os.walk(BASE_DIR):
        for file in files:
            full = os.path.join(root, file)
            z.write(full, arcname=os.path.relpath(full, BASE_DIR))

print("✅ ZIP READY:", zip_path)



```
